```python
import pandas as pd

data = pd.read_csv("./data.csv").dropna(axis="index", how="all").reset_index(drop=True)
data
```

```python
import json

name = "連小薇"
first_name = "小薇"

count = 0
with open("lumi_v0_6.jsonl", "w") as f:
    for i, item in data.iterrows():
        messages = []

        # system message
        messages.append({"role": "system", "content": system_message})

        # user / assistant messages
        for role, col in [
            ("user", "input"),
            ("assistant", "output"),
            ("user", "input2"),
            ("assistant", "output2"),
            ("user", "input3"),
            ("assistant", "output3"),
        ]:
            if pd.isna(item[col]):
                break
            message = (
                item[col].replace("{name}", name).replace("{first_name}", first_name)
            )
            messages.append({"role": role, "content": message})

        f.write(json.dumps({"messages": messages}) + "\n")
        count += 1

count
```
